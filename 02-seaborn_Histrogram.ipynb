### Import Dataset (mpg.csv)

```python
url='https://raw.githubusercontent.com/worachaiSM/tutorial/main/mpg.csv'
df=pd.read_csv(url)
df.head(5)
```

```python
sns.distplot(df['cty'] ,kde=True, label="City")
sns.distplot(df['hwy'] ,kde=True, label="HighWay")
plt.legend()
```

```python
sns.distplot(df['cty'] ,kde=True)
plt.axvline(x=df.cty.mean(), color='salmon', linestyle="-")
plt.title(f"Mean = {df.cty.mean():.2f}, S.D. ={df.cty.std():.4f}");
```
